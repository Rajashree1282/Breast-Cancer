# Breast Cancer Survival Data Set-EDA

Data Set Information: The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.



# Reference:
Breast Cancer Survival data set:- https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/Breast_cancer_survival.csv

Attribute Information:

1.Age:-Age of patient at time of operation (numerical)
2.Years_of_operation:-Patient's year of operation (year - 1900, numerical)
3.Pos_axillary_nodes:-Number of positive axillary nodes detected (numerical)
4.Status:-Survival status (class attribute) -- 1 = the patient survived 5 years or longer -- 2 = the patient died within 5 year

# Business Goal

To analyse each feature in Breast cancer survival dataset, and find the important features among available features (patient_age, year_of_operation, positive_axillary_nodes) that can be used to predict survival_status of a patient who had undergone surgery for breast cancer.

Q1: Which year does most of the operation occured for breast cancer?

```python
(Br_sur['Years_of_operation'].value_counts() / len(Br_sur)).plot(kind='bar', figsize=(12,7), colormap='magma', fontsize=12)
plt.xlabel('Years_of_operation', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title=('Bar Chart showing the Frequency in each Years_of_operation value')
plt.show()
```

Observation: 
Most of the operation occurred in 1958 for breast cancer.

Q2: Which age patients suffer from breast cancer?

```python
Br_sur['Age'].plot(kind='kde', figsize=(15,7), color='red')

plt.xlabel('Age', fontsize=15)
plt.ylabel('Status', fontsize=15)
plt.title=('KDE plot showing the survival rate of Age')
plt.show()
```

Observation:
Patients who suffer from breast cancer are of age 30 to 80

Q3: What is average of Pos_axillary_nodes in most of patients?

```python
Br_sur.groupby(['Pos_axillary_nodes'])['Status'].mean().sort_values().plot(kind='line', figsize=(15,7), color='blue', fontsize=13, linestyle='-.')
plt.xlabel('Pos_axillary_nodes', fontsize=12)
plt.ylabel('Status', fontsize=12)
plt.title=('Line chart showing the mean value of Pos_axillary_nodes with Status')
plt.show()
```

Obseravtion:
Most of the patients have 5 nodes on an average

Q4: In which age and year the number of patients are not survived?

```python
plt.figure(figsize=(12,7))
sns.boxplot(data=Br_sur, x='Status', y='Pos_axillary_nodes', palette='rainbow')
plt.title=('Status vs Pos_axillary_nodes')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(data=Br_sur, x='Status', y='Age', palette='rainbow')
plt.title=('Status vs Age')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(data=Br_sur, x='Status', y='Years_of_operation', palette='rainbow')
plt.title=('Status vs Years_of_operation')
plt.show()
```

Obesravtion:
The number of patients who are not survived have age between 46-62 and year between 1959-1965.

Q5: What is correlation between Age and Pos_axillary_node?

```python
#Using seaborn's pairplot function to plot the pairplot for specific columns of the dataset.
sns.pairplot(data=Br_sur[['Age', 'Years_of_operation', 'Pos_axillary_nodes']], size=3, diag_kind='kde')
plt.show()
```

From above obesravtion there is no correlation between Age and Pos_axillary_node.
But we see that:
1.Almost all the Patients are of age 30 - 80
2.Number of nodes many of the patients have are below 20

Q6:At what age most of patient dead within 5 years?
